```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df=pd.read_csv('/content/Mall_Customers.csv')
print(df.head(15))
```

```python
from sklearn.cluster import KMeans
x=df[['Age','Spending Score (1-100)']].copy()
```

Determine the number of Python clusters<br>
use the elbow method, which plots the within-clusters-sum-of-squares(wcss) versus the number of clusters

```python
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(x)
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title('Selecting the no. of clusters using the Elbow method')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
for k in range(1,11):
  plt.scatter(x["Age"],x["Spending Score (1-100)"])
  plt.title("Clusters identified by K-Means clustering")
plt.ylabel("Spending Score (1-100)")
plt.xlabel("Age")
plt.show()
```
