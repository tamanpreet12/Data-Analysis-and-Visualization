```python
import plotly.graph_objects as go
import pandas as pd
```

```python
country = pd.read_csv('countries.csv')
country.head()
```

```python
country.info()
```

```python
feature_to_visualize = 'Population'

fig = go.Figure(data=go.Choropleth(
    locations = country['Country'],
    locationmode = 'country names', # Assuming country names are used
    z = country[feature_to_visualize],
    text = country['Country'], # Hover text
    colorscale = 'Viridis', # Choose a colorscale
    colorbar_title = feature_to_visualize
))

fig.update_layout(
    title_text = f'Choropleth Map of {feature_to_visualize}',
    geo=dict(
        showcoastlines=True, # Show country coastlines
    )
)

fig.show()
```
