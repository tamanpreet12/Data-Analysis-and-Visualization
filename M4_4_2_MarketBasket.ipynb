```python
Market basket analysis is used by companies to identify items that are frequently purchased together.<br>
How does it work?<br>
It is frequently used by restaurants , retail stores, and online shopping platforms to encourage customers to make more purchases in a single visit.
This is a use-case of data science in marketing that increases company sales and drives business growth and commonly utilizes the Apriori algorithm.
```

Apriori algo. is the most common technique for performing market basket analysis.
It is used for association rule mining, which is a rule-based process used to identify correlations b/w items puchased by users.

Components of apriori:<br>
. support<br>
. lift<br>
.confidence<br>

Support<br>
The first component of apriori algo. is support- used to access the overall popularity of a given product with the following formula:<br>
Support(item)=Transaction comprising the item/Total transactions<br>


Confidence<br>
confidence(bread->milk)=transactions comprising bread and milk/ transactions comprising bread

Lift<br>
refers to the increase in the ratio of the sale of milk when you sell bread:<br>
lift=confidence(bread->milk)/support(bread)=0.75/1=1.3<br>
this means that customers are 1.3 times more likely to buy milk if you also sell bread

step 1: downlod dataset "groceries_dataset.csv"<br>
step 2: reading dataset

```python
import pandas as pd
from google.colab import drive

df=pd.read_csv('/content/Groceries_dataset.csv')
df.head()
```

```python
df['single_transaction']=df['Member_number'].astype(str)+'_'+df['Date'].astype(str)
df.head()
```

```python
df2=pd.crosstab(df['single_transaction'],df['itemDescription'])
df2.head()
```

```python
def encode(item_freq):
  res=0
  if item_freq>0:
    res=1
  return res
basket_input=df2.applymap(encode)
```

step 4: build the apriori algorithm

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
frequent_itemsets=apriori(basket_input,min_support=0.001,use_colnames=True)
rules=association_rules(frequent_itemsets,metric="lift")
rules.head()
```

```python
rules.sort_values(["support","confidence","lift"],axis=0,ascending=False).head(8)
```
